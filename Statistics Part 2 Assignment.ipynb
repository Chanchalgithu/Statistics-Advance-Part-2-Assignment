# Statistics Part 2

# 1. What is hypothesis testing in statistics ?

Hypothesis testing is a statistical method used to decide whether there is enough evidence in a sample to support a specific claim about a population.

 - Null Hypothesis (H₀): No effect or difference.

 - Alternative Hypothesis (H₁): There is an effect or difference.

 - Use sample data to calculate a test statistic and p-value.

 - Reject H₀ if p-value ≤ significance level (e.g., 0.05), otherwise fail to reject H₀.

It helps make data-driven decisions or judgments.

# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?

### Null Hypothesis vs Alternative Hypothesis

| Aspect                  | Null Hypothesis (H₀)                              | Alternative Hypothesis (H₁ or Ha)                      |
|-------------------------|--------------------------------------------------|--------------------------------------------------------|
| **Definition**          | A statement of **no effect** or **no difference** | A statement that **there is an effect** or **difference** |
| **Purpose**             | The starting assumption to be tested             | What you want to prove or provide evidence for         |
| **Example (Mean)**      | H₀: μ = 50 (mean is 50)                          | H₁: μ ≠ 50, μ > 50, or μ < 50 (mean is not 50)         |
| **Decision**            | Either **reject** or **fail to reject** H₀       | If H₀ is rejected, H₁ is **supported**                 |
| **Assumption**          | Assumes no change or relationship                | Suggests a new effect, change, or relationship         |

> ✅ In hypothesis testing, you collect sample data to decide whether to **reject** the null hypothesis in favor of the alternative.

# 3. What is the significance level in hypothesis testing, and why is it important ?

The **significance level** (denoted by **α**, alpha) is the threshold probability used to decide whether to **reject the null hypothesis (H₀)**.

---

### 🔹 Common Values:
- **α = 0.05** (5%) — most common
- **α = 0.01** (1%)
- **α = 0.10** (10%)

---

### 🔹 Why It Is Important:
- It defines the **maximum risk of making a Type I error** (rejecting a true null hypothesis).
- It sets the **standard for evidence** — how strong the sample data must be to reject H₀.
- Helps ensure decisions are not made due to random chance alone.

---

### 🔹 Example:
If **α = 0.05**, you are willing to accept a **5% chance of wrongly rejecting** a true null hypothesis.

> ✅ A lower α means stricter criteria (more confidence needed), while a higher α increases the chance of detecting an effect but also the risk of a false alarm.

# 4.  What does a P-value represent in hypothesis testing ?

The **P-value** is the **probability** of obtaining the observed result (or something more extreme) **assuming the null hypothesis (H₀) is true**.

---

### 🔹 Interpretation:
- **Small P-value (≤ α):** Strong evidence **against** H₀ → **Reject H₀**
- **Large P-value (> α):** Weak evidence against H₀ → **Fail to reject H₀**

---

### 🔹 Example:
If **P = 0.03** and **α = 0.05**,  
→ There is a 3% chance of observing the data if H₀ is true.  
→ Since 0.03 < 0.05, you **reject H₀**.

---

### 🔹 Key Points:
- P-value **does not** tell you the probability that H₀ is true.
- It helps you measure the **strength of evidence** against H₀.

> ✅ The smaller the P-value, the stronger the evidence against the null hypothesis.

# 5. How do you interpret the P-value in hypothesis testing ?

### How to Interpret the P-value in Hypothesis Testing

The **P-value** helps you decide whether to **reject the null hypothesis (H₀)** based on how likely your sample result is, assuming H₀ is true.

---

### 🔹 General Interpretation:

- **P ≤ α (significance level):**
  - The result is **statistically significant**.
  - There is **strong evidence against H₀**.
  - 👉 **Reject H₀**

- **P > α:**
  - The result is **not statistically significant**.
  - There is **not enough evidence to reject H₀**.
  - 👉 **Fail to reject H₀**

---

### 🔹 Example:
If **P = 0.02** and **α = 0.05**:  
- 2% chance of getting such a result if H₀ is true.  
- Since 0.02 < 0.05, reject H₀ — the result is significant.

---

### 🔹 Reminder:
- A **smaller P-value** = **stronger evidence** against H₀.
- A **larger P-value** = **weaker evidence** against H₀.
- P-value **does not** indicate the probability that H₀ is true or false.

> ✅ Use the P-value to **quantify the strength of evidence** in your hypothesis test.
    
# 6. What are Type 1 and Type 2 errors in hypothesis testing ?

### Type I and Type II Errors in Hypothesis Testing

When making decisions based on hypothesis tests, there are two possible types of errors:

---

### 🔹 Type I Error (False Positive):
- **Definition:** Rejecting the **null hypothesis (H₀)** when it is actually **true**.
- **Symbol:** α (alpha)
- **Example:** Concluding a medicine works when it actually doesn't.

✅ **You detect an effect that isn't really there.**

---

### 🔹 Type II Error (False Negative):
- **Definition:** Failing to reject the **null hypothesis (H₀)** when the **alternative hypothesis (H₁)** is actually **true**.
- **Symbol:** β (beta)
- **Example:** Concluding a medicine doesn't work when it actually does.

❌ **You miss a real effect.**

---

### 🔹 Summary Table:

| Error Type     | What Happens                        | Real Truth        | Your Decision           |
|----------------|-------------------------------------|-------------------|--------------------------|
| **Type I (α)** | Reject H₀ when it's actually true   | H₀ is true        | ❌ Wrong (False Positive) |
| **Type II (β)**| Fail to reject H₀ when it's false   | H₁ is true        | ❌ Wrong (False Negative) |

> 🎯 **Goal:** Minimize both errors, but there’s always a trade-off. Lowering one can increase the other.

# 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?

### One-Tailed vs Two-Tailed Test in Hypothesis Testing

---

### 🔹 One-Tailed Test:
- Tests for an effect in **one specific direction** (greater than or less than).
- **Alternative Hypothesis (H₁)**:  
  - H₁: μ > μ₀ → right-tailed test  
  - H₁: μ < μ₀ → left-tailed test
- **Used when** you only care about deviation in **one direction**.

**Example:**  
H₀: μ = 100  
H₁: μ > 100 → You only test if the mean is **greater than** 100.

---

### 🔹 Two-Tailed Test:
- Tests for an effect in **both directions** (not equal).
- **Alternative Hypothesis (H₁):**  
  - H₁: μ ≠ μ₀
- **Used when** you want to detect **any significant difference**, either higher or lower.

**Example:**  
H₀: μ = 100  
H₁: μ ≠ 100 → You test if the mean is **different** from 100 in **either direction**.

---

### 🔹 Comparison Table:

| Feature              | One-Tailed Test                | Two-Tailed Test               |
|----------------------|-------------------------------|-------------------------------|
| **Direction**        | One direction (>, <)          | Both directions (≠)           |
| **Alternative H₁**   | μ > μ₀ or μ < μ₀              | μ ≠ μ₀                        |
| **Use Case**         | Specific directional claim     | General difference            |
| **Rejection Region** | Only on one tail               | Both tails                    |
| **More Powerful?**   | More power in one direction    | Balanced test in both         |

> ✅ Choose the test based on the hypothesis and research question you are testing.

# 8. What is the Z-test, and when is it used in hypothesis testing ?     

### What is the Z-test and When is It Used?

---

### 🔹 What is a Z-test?

A **Z-test** is a type of hypothesis test used to determine whether there is a significant difference between sample and population means (or between two sample means) **when the population standard deviation is known** and the sample size is large (n ≥ 30).

It uses the **Z-distribution (standard normal distribution)**.

---

### 🔹 Z-test Formula:

For one sample mean:

\[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
\]

Where:  
- \(\bar{X}\) = sample mean  
- \(\mu\) = population mean  
- \(\sigma\) = population standard deviation  
- \(n\) = sample size

---

### 🔹 When to Use the Z-test:

✅ Use the Z-test when:  
- The **sample size is large** (n ≥ 30)  
- The **population standard deviation (σ)** is **known**  
- The data is **approximately normally distributed**

---

### 🔹 Types of Z-tests:

1. **One-Sample Z-test:**  
   Compare sample mean to a known population mean.

2. **Two-Sample Z-test:**  
   Compare means of two independent large samples.

3. **Z-test for Proportions:**  
   Test differences between population proportions.

---

### 🔹 Example:

A company claims the average battery life is 100 hours.  
You test a sample of 50 batteries and want to know if the mean is significantly different from 100 hours.  
If population standard deviation is known → **Z-test** is appropriate.

> ✅ The Z-test is fast, simple, and ideal when conditions are met. Otherwise, use a **T-test**.

# 9.  How do you calculate the Z-score, and what does it represent in hypothesis testing ?


---

### 🔹 What is a Z-score?

A **Z-score** measures how many standard deviations a data point (or sample mean) is away from the population mean under the null hypothesis.

It tells you the **position of the sample result relative to the population mean**.

---

### 🔹 Z-score Formula:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:  
- \(X\) = observed value (data point or sample mean)  
- \(\mu\) = population mean  
- \(\sigma\) = population standard deviation

---

### 🔹 In Hypothesis Testing:

- The Z-score helps calculate the **test statistic**.
- It shows how extreme the sample result is if the null hypothesis (H₀) is true.
- You compare the Z-score with critical values from the **standard normal distribution** to decide whether to reject H₀.

---

### 🔹 Example:

If a sample mean is 105, population mean is 100, and population standard deviation is 5:

\[
Z = \frac{105 - 100}{5} = 1
\]

This means the sample mean is 1 standard deviation above the population mean.

---

### 🔹 Interpretation:

- Larger absolute Z-scores mean more evidence **against H₀**.
- Smaller absolute Z-scores mean less evidence **against H₀**.

> ✅ The Z-score quantifies how unusual the sample result is under the assumption that H₀ is true.

# 10. What is the T-distribution, and when should it be used instead of the normal distribution ?

### What is the T-distribution, and When to Use It?

---

### 🔹 What is the T-distribution?

- The **T-distribution** (or Student’s t-distribution) is a probability distribution similar to the normal distribution but with **heavier tails**.
- It accounts for extra uncertainty when the **population standard deviation (σ) is unknown** and must be estimated from the sample.
- It depends on the **degrees of freedom (df)**, usually \(df = n - 1\), where \(n\) is the sample size.
- As the sample size increases, the t-distribution approaches the normal distribution.

---

### 🔹 When to Use the T-distribution Instead of the Normal Distribution:

✅ Use the **T-distribution** when:  
- The **sample size is small** (typically \(n < 30\)).  
- The **population standard deviation (σ) is unknown**.  
- The data is approximately **normally distributed** or close to normal.

---

### 🔹 Why Use T-distribution?

- Because estimating σ from a small sample adds uncertainty, the t-distribution’s heavier tails give a more accurate test result.
- It reduces the risk of **Type I errors** due to this uncertainty.

---

### 🔹 Summary Table:

| Condition                     | Use Z-test (Normal)                 | Use T-test (T-distribution)             |
|-------------------------------|----------------------------------|----------------------------------------|
| Population standard deviation | Known                           | Unknown                                |
| Sample size                   | Large (n ≥ 30)                  | Small (n < 30)                        |
| Distribution assumption       | Normal or large sample size     | Approximately normal                   |

> ✅ When in doubt with small samples and unknown σ, prefer the **T-distribution**.

# 12. What is the T-test, and how is it used in hypothesis testing ?

### 🔹 What is a T-test?

The **T-test** is a statistical test used to determine if there is a significant difference between the means of two groups, especially when:  
- The **population standard deviation (σ) is unknown**  
- The **sample size is small** (usually \(n < 30\))  

It uses the **Student’s t-distribution** to calculate the test statistic.

---

### 🔹 Types of T-tests:

1. **One-sample T-test:**  
   Tests if the sample mean is different from a known or hypothesized population mean.

2. **Independent two-sample T-test:**  
   Compares means of two independent groups.

3. **Paired sample T-test:**  
   Compares means from the same group at different times or under different conditions.

---

### 🔹 T-test Formula (One-Sample):

\[
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
\]

Where:  
- \(\bar{X}\) = sample mean  
- \(\mu\) = population mean (under H₀)  
- \(s\) = sample standard deviation  
- \(n\) = sample size

---

### 🔹 How is it Used in Hypothesis Testing?

- Formulate null hypothesis \(H_0\) and alternative hypothesis \(H_1\).
- Calculate the t-statistic using sample data.
- Determine the critical t-value from the t-distribution table (based on degrees of freedom and significance level).
- Compare the calculated t-statistic to the critical value:
  - If \(|t|\) > critical value → **Reject \(H_0\)**  
  - Else → **Fail to reject \(H_0\)**

---

> ✅ The T-test helps decide if the observed sample difference is statistically significant, accounting for uncertainty when σ is unknown.

# 13. What is the relationship between Z-test and T-test in hypothesis testing ?

### Relationship Between Z-test and T-test in Hypothesis Testing

---

### 🔹 Common Purpose:
- Both tests are used to compare sample means and test hypotheses about population means.

---

### 🔹 Key Relationship:
- The **T-test** is a generalization of the **Z-test**.
- When the **population standard deviation (σ) is known** and the **sample size is large (n ≥ 30)**, the **T-distribution approximates the Normal (Z) distribution**.
- In this case, the **T-test and Z-test give very similar results**.
- When **σ is unknown** and/or the **sample size is small (n < 30)**, the **T-test** should be used because it accounts for extra uncertainty with heavier tails in the distribution.

---

### 🔹 Summary:

| Condition                        | Use Z-test                  | Use T-test                   |
|---------------------------------|-----------------------------|-----------------------------|
| Population standard deviation   | Known                       | Unknown                     |
| Sample size                     | Large (≥ 30)                | Small (< 30)                |
| Distribution                   | Normal (Z-distribution)     | Student’s t-distribution     |
| Approximation                  | T-test approaches Z-test as sample size grows |

---

> ✅ **In short:** The T-test extends the Z-test to situations with unknown population variance and small samples, but they converge as sample size grows.

# 14. What is a confidence interval, and how is it used to interpret statistical results ?

### What is a Confidence Interval, and How is it Used?

---

### 🔹 What is a Confidence Interval (CI)?

A **confidence interval** is a range of values, derived from sample data, that is likely to contain the **true population parameter** (like the mean or proportion) with a certain level of confidence.

---

### 🔹 Confidence Level:

- Common levels: **90%**, **95%**, **99%**
- A **95% confidence interval** means:
  > "We are 95% confident that the true population value lies within this interval."

---

### 🔹 CI Formula (for mean when σ is known):

\[
\text{CI} = \bar{X} \pm Z_{\alpha/2} \cdot \left( \frac{\sigma}{\sqrt{n}} \right)
\]

Where:  
- \(\bar{X}\) = sample mean  
- \(Z_{\alpha/2}\) = critical Z-value  
- \(\sigma\) = population standard deviation  
- \(n\) = sample size

*(Use t-value and sample standard deviation if σ is unknown.)*

---

### 🔹 How It's Used in Interpretation:

- If a **95% CI** for the mean is [48, 52], we can say:  
  > “We are 95% confident the population mean is between 48 and 52.”

- If a **confidence interval does not include** a hypothesized value (like 0 or some mean μ₀), it may **support rejecting the null hypothesis**.

---

### 🔹 Example in Hypothesis Testing:

- Suppose you're testing \( H_0: \mu = 100 \)
- Your 95% CI is [105, 110]
- Since **100 is not in the interval**, you may **reject \( H_0 \)**

---

> ✅ **Conclusion:** Confidence intervals provide both an estimate and a margin of error. They help assess the reliability of sample statistics and guide decision-making in hypothesis testing.

# 15. What is the margin of error, and how does it affect the confidence interval ?

### 🔹 What is the Margin of Error (MoE)?

The **margin of error** is the amount added and subtracted from the sample statistic (like the mean or proportion) to create a **confidence interval**.

It represents the **range of uncertainty** due to sampling variability.

---

### 🔹 Formula for Margin of Error:

For population mean (with known σ):

\[
\text{MoE} = Z_{\alpha/2} \cdot \left( \frac{\sigma}{\sqrt{n}} \right)
\]

Where:  
- \(Z_{\alpha/2}\) = critical Z-value (based on confidence level)  
- \(\sigma\) = population standard deviation  
- \(n\) = sample size

*(Use t-value and sample standard deviation if σ is unknown.)*

---

### 🔹 How It Affects the Confidence Interval:

The **confidence interval (CI)** is calculated as:

\[
\text{CI} = \bar{X} \pm \text{MoE}
\]

So:
- A **larger margin of error** means a **wider confidence interval**, indicating **less precision**.
- A **smaller margin of error** means a **narrower confidence interval**, indicating **more precision**.

---

### 🔹 Factors That Influence Margin of Error:

| Factor            | Effect on MoE                    |
|-------------------|----------------------------------|
| **Sample size (n)** | ↑ n → ↓ MoE (more precise)      |
| **Confidence level** | ↑ confidence → ↑ MoE (wider CI) |
| **Standard deviation (σ)** | ↑ σ → ↑ MoE (more variability)  |

---

### 🔹 Example:

If sample mean = 70 and MoE = 5,  
then 95% CI = [65, 75]

---

> ✅ **Conclusion:** The margin of error determines how "wide" your confidence interval is, directly impacting how precise your estimate is.

# 16. How is Bayes' Theorem used in statistics, and what is its significance ?

### 🔹 What is Bayes' Theorem?

**Bayes' Theorem** is a formula that describes how to update the **probability of a hypothesis** based on **new evidence**.

---

### 🔹 Formula:

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

Where:  
- \(P(A|B)\) = Posterior probability (probability of A given B)  
- \(P(B|A)\) = Likelihood (probability of B given A)  
- \(P(A)\) = Prior probability of A  
- \(P(B)\) = Total probability of B

---

### 🔹 How It's Used in Statistics:

1. **Updating Beliefs:**  
   - Start with an initial belief (prior), and update it when new data (evidence) is observed.
   - Especially useful in **medical testing**, **machine learning**, and **spam filtering**.

2. **Classification Models:**  
   - Forms the basis of **Naive Bayes Classifiers** used in data science.

3. **Decision-Making Under Uncertainty:**  
   - Helps in **probabilistic reasoning**, e.g., determining the likelihood of an event after observing new outcomes.

---

### 🔹 Example:

In medical testing:

- Let A = person has a disease  
- Let B = test result is positive  
Bayes’ theorem helps find the **true probability** that the person has the disease **given a positive test result**.

---

### 🔹 Significance:

✅ Allows **reversal of conditional probabilities**  
✅ Combines **prior knowledge** with **new evidence**  
✅ Key concept in **Bayesian statistics** and **real-world decision-making**

---

> **Conclusion:**  
Bayes’ Theorem is a powerful tool for reasoning under uncertainty. It allows statisticians to update probabilities as more data becomes available, making it crucial in modern analytics and AI.

# 17. What is the Chi-square distribution, and when is it used ?

### 🔹 What is the Chi-square Distribution?

- The **Chi-square (χ²) distribution** is a continuous probability distribution.
- It is the distribution of the **sum of squared standard normal variables**.
- It is **positively skewed** and only takes **non-negative values** (χ² ≥ 0).
- Its shape depends on the **degrees of freedom (df)**.

---

### 🔹 Formula (Conceptual):

If \( Z_1, Z_2, ..., Z_k \) are independent standard normal variables, then:

\[
\chi^2 = Z_1^2 + Z_2^2 + \cdots + Z_k^2
\]

---

### 🔹 When Is It Used?

Chi-square distribution is mainly used in **hypothesis testing** involving **categorical data**:

1. **Chi-square Test of Independence**  
   - Checks whether two categorical variables are independent or related.  
   - Example: Is gender related to product preference?

2. **Chi-square Goodness-of-Fit Test**  
   - Tests if an observed frequency distribution matches an expected distribution.  
   - Example: Does a die roll uniformly?

3. **Test for Population Variance**  
   - When testing if the variance of a population equals a specific value.

---

### 🔹 Properties:

| Property                | Value                        |
|-------------------------|------------------------------|
| Range                   | \( \chi^2 \geq 0 \)          |
| Shape                   | Skewed right (becomes normal as df ↑) |
| Mean                    | Equal to degrees of freedom (df) |
| Variance                | \( 2 \times \text{df} \)     |

---

> ✅ **Conclusion:**  
The Chi-square distribution is essential for analyzing **categorical data** and testing **independence** or **goodness-of-fit** in statistics.                                 
                                 
# 18. What is the Chi-square goodness of fit test, and how is it applied ?

### 🔹 What Is It?

The **Chi-square Goodness-of-Fit Test** is a statistical method used to determine whether an observed frequency distribution matches an expected (theoretical) distribution.

---

### 🔹 Purpose:

To test if a **categorical variable** follows a **specific distribution** (e.g., uniform, binomial, etc.).

---

### 🔹 When to Use:

- You have **one categorical variable**
- You want to test how well the observed data fits an **expected distribution**

---

### 🔹 Hypotheses:

- **Null Hypothesis \(H_0\)**: The observed data follows the expected distribution.
- **Alternative Hypothesis \(H_1\)**: The observed data does not follow the expected distribution.

---

### 🔹 Formula:

\[
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
\]

Where:  
- \(O_i\) = observed frequency for category *i*  
- \(E_i\) = expected frequency for category *i*

---

### 🔹 Steps to Apply:

1. **State hypotheses**: \(H_0\) and \(H_1\)
2. **Calculate expected frequencies** based on the assumed distribution.
3. **Compute the Chi-square statistic** using the formula above.
4. **Find degrees of freedom**:  
   \[
   \text{df} = k - 1
   \]  
   where *k* = number of categories
5. **Compare** the computed χ² value to the critical value from the Chi-square table or use the **P-value**.
6. **Decision**:  
   - If χ² is large or **P-value < significance level**, reject \(H_0\).

---

### 🔹 Example:

| Category | Observed (O) | Expected (E) |
|----------|--------------|--------------|
| A        | 30           | 25           |
| B        | 20           | 25           |
| C        | 50           | 50           |

\[
\chi^2 = \frac{(30-25)^2}{25} + \frac{(20-25)^2}{25} + \frac{(50-50)^2}{50} = 2.0
\]

Then compare with critical χ² value at df = 2 and desired α (e.g., 0.05).

---

> ✅ **Conclusion:**  
The Chi-square goodness-of-fit test helps check if the observed frequencies significantly differ from what is expected under a certain theoretical distribution.

# 19. What is the F-distribution, and when is it used in hypothesis testing ?

### 🔹 What is the F-distribution?

- The **F-distribution** is a continuous probability distribution.
- It arises as the ratio of **two independent chi-square distributions** divided by their respective degrees of freedom.
- It is **skewed to the right** and used when comparing **variances**.

---

### 🔹 Formula:

\[
F = \frac{(\chi^2_1 / df_1)}{(\chi^2_2 / df_2)}
\]

Where:  
- \(\chi^2_1, \chi^2_2\) are chi-square variables  
- \(df_1, df_2\) are their degrees of freedom

---

### 🔹 When Is It Used?

The **F-distribution** is used in hypothesis testing when comparing **variances** or evaluating **multiple means**:

1. **ANOVA (Analysis of Variance)**  
   - To compare **3 or more group means** to see if at least one is different  
   - H₀: All group means are equal  
   - H₁: At least one group mean is different

2. **F-test for Equality of Variances**  
   - To test if **two populations have equal variances**

3. **Regression Analysis**  
   - To test the overall **significance of a regression model**

---

### 🔹 Properties:

| Property         | Value                                |
|------------------|----------------------------------------|
| Range            | \(F \geq 0\)                          |
| Shape            | Right-skewed                         |
| Depends on       | Two sets of degrees of freedom: \(df_1, df_2\) |
| Mean (if df₂ > 2)| \(\frac{df_2}{df_2 - 2}\)             |

---

### 🔹 Example (ANOVA context):

If you're testing the average test scores of **3 classes**, the F-distribution helps determine if at least one class average is **significantly different** from the others.

---

> ✅ **Conclusion:**  
The F-distribution is crucial in statistical tests that involve **comparing variances or multiple means**, such as ANOVA and regression analysis.    

# 20.  What is an ANOVA test, and what are its assumptions ?

### 🔹 What is ANOVA?

**ANOVA** stands for **Analysis of Variance**.  
It is a statistical method used to **compare the means of three or more groups** to determine if at least one group mean is significantly different from the others.

---

### 🔹 Purpose:

- Tests the null hypothesis \(H_0\): All group means are equal.
- Alternative hypothesis \(H_1\): At least one group mean is different.

---

### 🔹 Formula (F-statistic):

\[
F = \frac{\text{Between-group variability}}{\text{Within-group variability}} = \frac{MS_{\text{between}}}{MS_{\text{within}}}
\]

Where:
- \(MS = \frac{SS}{df}\) (Mean Square = Sum of Squares / Degrees of Freedom)

---

### 🔹 Types of ANOVA:

1. **One-way ANOVA** – compares means of **one factor** across multiple groups.  
2. **Two-way ANOVA** – compares means with **two independent variables**.

---

### 🔹 Assumptions of ANOVA:

1. **Independence of observations**  
   - Each group sample must be independent from the others.

2. **Normality**  
   - The data in each group should be approximately **normally distributed**.

3. **Homogeneity of variance (equal variances)**  
   - All groups should have **roughly equal variances**.

---

### 🔹 Example:

Comparing average scores of 3 different teaching methods:
- H₀: Mean₁ = Mean₂ = Mean₃
- H₁: At least one mean differs

Use ANOVA to check if the teaching method has a significant effect on scores.

---

> ✅ **Conclusion:**  
ANOVA is a powerful tool to test differences among **multiple means**. It relies on key assumptions like **independent observations, normal distribution, and equal variances**.

# 21. What are the different types of ANOVA tests ?

### 🔹 1. One-Way ANOVA

- **Purpose**: Compares the means of **three or more groups** based on **one independent variable (factor)**.
- **Example**: Compare average test scores across 3 teaching methods.
- **Hypotheses**:
  - H₀: μ₁ = μ₂ = μ₃ (all means are equal)
  - H₁: At least one mean is different

---

### 🔹 2. Two-Way ANOVA (Factorial ANOVA)

- **Purpose**: Compares means across groups based on **two independent variables (factors)**.
- Tests:
  - Main effect of Factor A
  - Main effect of Factor B
  - Interaction effect between A and B
- **Example**: Study the effect of teaching method and gender on student performance.

---

### 🔹 3. Repeated Measures ANOVA

- **Purpose**: Used when **same subjects** are tested **multiple times** (repeated observations).
- **Example**: Measure blood pressure of the same patients over 3 weeks.
- Reduces variability since subjects act as their own control.

---

### 🔹 4. MANOVA (Multivariate ANOVA)

- **Purpose**: Extension of ANOVA when there are **multiple dependent variables**.
- **Example**: Compare students from 3 schools on both **math and reading scores**.
- Tests for group differences in a **combination of dependent variables**.

---

### 🔹 5. ANCOVA (Analysis of Covariance)

- **Purpose**: Like ANOVA but includes **one or more covariates** to control for external variables.
- **Example**: Compare student performance across schools while controlling for **socioeconomic status**.

---

> ✅ **Summary Table:**

| Type               | Factors | Dependent Variables | Repeats | Covariates |
|--------------------|---------|----------------------|---------|------------|
| One-Way ANOVA       | 1       | 1                    | No      | No         |
| Two-Way ANOVA       | 2       | 1                    | No      | No         |
| Repeated Measures   | 1       | 1                    | Yes     | No         |
| MANOVA              | 1+      | 2+                   | No/Yes  | No         |
| ANCOVA              | 1+      | 1                    | No      | Yes        |

---

> 📝 **Conclusion:**  
Each ANOVA type serves a specific research scenario, from comparing group means to analyzing multiple variables or adjusting for covariates.

# 22. What is the F-test, and how does it relate to hypothesis testing ?

### 🔹 What is the F-test?

The **F-test** is a statistical test that compares **two variances** or evaluates the **overall significance of a model**, such as in ANOVA or regression.

It uses the **F-distribution** and is based on the ratio of two variances:

\[
F = \frac{\text{Variance}_1}{\text{Variance}_2}
\]

---

### 🔹 Purpose of the F-test:

1. **Compare Variances**  
   - To test if two populations have equal variances.

2. **ANOVA (Analysis of Variance)**  
   - To check if **three or more group means** are significantly different.

3. **Regression Analysis**  
   - To assess if the overall regression model is statistically significant.

---

### 🔹 Hypotheses (for comparing variances):

- **Null Hypothesis \(H_0\)**: Variances are equal \((\sigma_1^2 = \sigma_2^2)\)  
- **Alternative Hypothesis \(H_1\)**: Variances are not equal \((\sigma_1^2 \ne \sigma_2^2)\)

---

### 🔹 Decision Rule:

- Calculate the **F-statistic**
- Find the **critical F-value** from the F-distribution table using degrees of freedom.
- If:
  \[
  F_{\text{calc}} > F_{\text{critical}}, \quad \text{reject } H_0
  \]

Or use the **P-value** method:
- If \( \text{P-value} < \alpha \), reject \(H_0\)

---

### 🔹 Relationship to Hypothesis Testing:

The F-test is a **core component** of hypothesis testing when:
- Testing **equality of variances**
- Evaluating **multiple group means** (ANOVA)
- Validating **overall model fit** in regression

---

> ✅ **Conclusion:**  
The F-test is essential in hypothesis testing whenever you compare **variances** or test **multiple means**. It helps determine if observed differences are **statistically significant**.

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

# Sample data
sample = [68, 72, 65, 70, 74, 69, 71, 73, 67, 66]  # Sample values
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

# Population parameters
population_mean = 70   # Known population mean
population_std = 3     # Known population standard deviation (σ)
alpha = 0.05           # Significance level

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output results
print("Sample Mean:", sample_mean)
print("Z-score:", round(z_score, 3))
print("P-value:", round(p_value, 4))

# Interpretation
if p_value < alpha:
    print("👉 Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("✅ Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

```python
# 5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing. 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    se = sigma / np.sqrt(n)
    z_critical = norm.ppf(1 - alpha)
    x = np.linspace(-4, 4 + mu1, 1000)

    # Null and alternative distributions
    null_dist = norm(loc=mu0, scale=se)
    alt_dist = norm(loc=mu1, scale=se)

    # Plot
    plt.plot(x, null_dist.pdf(x), label="H0 (null)", color='blue')
    plt.plot(x, alt_dist.pdf(x), label="H1 (alt)", color='green')

    # Critical value
    x_crit = mu0 + z_critical * se
    plt.axvline(x_crit, color='red', linestyle='--', label='Critical value')

    # Shade Type I Error (alpha)
    x_alpha = np.linspace(x_crit, mu0 + 4*se, 200)
    plt.fill_between(x_alpha, null_dist.pdf(x_alpha), color='red', alpha=0.3, label='Type I error (α)')

    # Shade Type II Error (beta)
    x_beta = np.linspace(mu0 - 4*se, x_crit, 200)
    plt.fill_between(x_beta, alt_dist.pdf(x_beta), color='orange', alpha=0.3, label='Type II error (β)')

    plt.title("Type I and Type II Errors")
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_errors()
```

```python
# 9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)
    margin = stats.t.ppf((1 + confidence) / 2, n - 1) * std_err
    return (mean - margin, mean + margin)

# Example usage
data = [12, 15, 14, 10, 13]
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_err = stats.sem(sample)
    t_score = stats.t.ppf((1 + confidence) / 2, n - 1)
    return t_score * std_err

# Example usage
data = [50, 52, 48, 47, 51, 49, 53]
moe = margin_of_error(data)
print(f"Margin of Error (95% CI): {moe:.2f}")
```

```python
# 16.  Implement an F-test using Python to compare the variances of two random samples. 

import numpy as np
from scipy.stats import f

# Sample data
np.random.seed(0)
sample1 = np.random.normal(10, 5, 30)
sample2 = np.random.normal(10, 3, 30)

# Calculate variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Calculate F-statistic (larger variance / smaller variance)
F = var1 / var2 if var1 > var2 else var2 / var1

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Calculate p-value (two-tailed)
p_value = 2 * min(f.cdf(F, df1, df2), 1 - f.cdf(F, df1, df2))

print(f"F-statistic: {F:.3f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Result: Variances are significantly different (reject H0)")
else:
    print("Result: No significant difference in variances (fail to reject H0)")
```

```python
# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results. 

import numpy as np
from scipy.stats import chi2

# Sample data
data = np.array([12, 15, 14, 10, 13, 16, 14, 15, 11, 13])

# Sample variance (unbiased)
s2 = np.var(data, ddof=1)

# Sample size
n = len(data)

# Hypothesized population variance
sigma2_0 = 4  # for example

# Calculate Chi-square statistic
chi2_stat = (n - 1) * s2 / sigma2_0

# Degrees of freedom
df = n - 1

# Significance level
alpha = 0.05

# Calculate critical values for two-tailed test
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)

# Calculate p-value
p_value = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))

print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"Critical values: {chi2_lower:.2f}, {chi2_upper:.2f}")
print(f"P-value: {p_value:.3f}")

if chi2_stat < chi2_lower or chi2_stat > chi2_upper:
    print("Result: Reject H0 - population variance is significantly different")
else:
    print("Result: Fail to reject H0 - no significant difference in variance")
```

```python
# 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups. 

from statsmodels.stats.proportion import proportions_ztest

# Number of successes in each group
successes = [45, 30]

# Number of observations in each group
n = [100, 100]

# Perform two-proportion Z-test
z_stat, p_value = proportions_ztest(successes, n)

print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Result: Proportions are significantly different (reject H0)")
else:
    print("Result: No significant difference in proportions (fail to reject H0)")
```

```python
# 26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results. 

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Sample data
np.random.seed(0)
data1 = np.random.normal(50, 5, 30)
data2 = np.random.normal(50, 8, 30)

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Calculate F-statistic (larger variance / smaller variance)
if var1 > var2:
    F = var1 / var2
    dfn = len(data1) - 1  # numerator df
    dfd = len(data2) - 1  # denominator df
else:
    F = var2 / var1
    dfn = len(data2) - 1
    dfd = len(data1) - 1

# Calculate p-value for two-tailed test
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

print(f"F-statistic: {F:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Result: Variances are significantly different (reject H0)")
else:
    print("Result: No significant difference in variances (fail to reject H0)")

# Visualization
x = np.linspace(0, 5, 500)
y = f.pdf(x, dfn, dfd)
plt.plot(x, y, label='F-distribution PDF')
plt.axvline(F, color='red', linestyle='--', label=f'F-statistic = {F:.2f}')
plt.title('F-distribution with Observed F-statistic')
plt.xlabel('F value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results. 

import numpy as np
from scipy.stats import chisquare

# Simulated observed frequencies (e.g., counts in 4 categories)
observed = np.array([50, 30, 10, 10])

# Expected frequencies (e.g., equal distribution)
expected = np.array([25, 25, 25, 25])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Result: Observed distribution differs significantly from expected (reject H0)")
else:
    print("Result: No significant difference from expected distribution (fail to reject H0)")
```
